```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../datasets/Fertilizer_Prediction.csv")
```

```python
df
```

```python
df['Soil Type'].unique()
```

```python
df['Crop Type'].unique()
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['Soil Type']= label_encoder.fit_transform(df['Soil Type'])
```

```python
df['Crop Type']= label_encoder.fit_transform(df['Crop Type'])
```

```python
X = df[['Temparature','Humidity','Moisture','Soil Type','Crop Type','Nitrogen','Potassium','Phosphorous']]
target = df['Fertilizer_Name']
```

```python
y = target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 30)
```

```python
print("X_train shape:",X_train.shape)
print("X_test shape:",X_test.shape)
print("y_train shape:",y_train.shape)
print("y_test shape:",y_test.shape)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1, max_depth=5,n_estimators=100, oob_score=True)
```

```python
%%time
rf.fit(X_train, y_train)
```

```python
rf.oob_score_
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred) 
```
